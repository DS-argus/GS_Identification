#### Examples in [Tian, Pearl 2003]

```python
# Figure 1: Front-door case
display(Image(filename='Images/03_Figure1.png'))

G = CD({'X', 'Z', 'Y'}, 
                    [('X', 'Z'), ('Z', 'Y')],
                    [('X', 'Y', 'U_XY')])

P1 = myID(Y={'Y'}, X={'X'}, G=G)
print(f"myID")
display(Math(P1.printLatex()))

P2 = mygID(Y=set(["Y"]), X=set(["X"]), Z=frozenset([frozenset()]), G=G)
print(f"mygID")
display(Math(P2.printLatex()))
```

```python
# Figure 2 --> 정리하면 논문 결과와 일치
display(Image(filename='Images/03_Figure2.png'))

G = CD({'X', 'Z1', 'Z2', 'Z3', 'Y'}, 
                    [('X', 'Z1'), ('Z1', 'Y'), ('Z2', 'X'), ('Z2', 'Z1'), ('Z2', 'Z3'), ('Z3', 'Y')],
                    [('X', 'Z2', 'U_XZ2'), ('Y', 'Z2', 'U_YZ2'), ('X', 'Z3', 'U_XZ3'), ('X', 'Y', 'U_XY')])

P1 = myID(Y={'Y', 'Z1', 'Z2', 'Z3'}, X={'X'}, G=G)
print(f"myID")
display(Math(P1.printLatex()))

P2 = mygID(Y={'Y', 'Z1', 'Z2', 'Z3'}, X={'X'}, Z=[set()], G=G)
print(f"mygID")
display(Math(P2.printLatex()))
```

```python
# Figure 5
display(Image(filename='Images/03_Figure5.png'))

G = CD({'X', 'Z', 'Y'}, 
                    [('X', 'Y')],
                    [('X', 'Z', 'U_XZ'), ('Z', 'Y', 'U_ZY')])

try:
    P1 = myID(Y={'Y', 'Z'}, X={'X'}, G=G)
    print(f"myID")
    display(Math(P1.printLatex()))
except HedgeFound as H:
    print(f"myID Failed: {H._message}")

try:
    P2 = mygID(Y={'Y', 'Z'}, X={'X'}, Z=[set()], G=G)
    print(f"gID")
    display(Math(P2.printLatex()))
except ThicketFound as T: 
    print(f"mygID Failed: {T._message}")
```

#### Examples in [Shpitser, Pearl 2006]

```python
# Figure 1.(a)
display(Image(filename='Images/06_Figure1a.png'))

G = CD({'W1', 'W2', 'X', 'Y1', 'Y2'}, 
                    [('W1', 'X'), ('X', 'Y1'), ('W2', 'Y2')],
                    [('W1', 'W2', 'U_W1W2'), ('W1', 'Y2', 'U_W1Y2'), ('W2', 'X', 'U_W2X'), ('W1', 'Y1', 'U_W1Y1')])

P = myID(Y={'Y1', 'Y2'}, X={'X'}, G=G)
print(f"myID")
display(Math(P.printLatex()))

P2 = mygID(Y={'Y1', 'Y2'}, X={'X'}, Z=[set()], G=G)
print(f"mygID")
display(Math(P2.printLatex()))
```

```python
# Figure 1.(b)
display(Image(filename='Images/06_Figure1b.png'))

G = CD({'W1', 'W2', 'X', 'Y1', 'Y2'}, 
                    [('W1', 'W2'), ('W1', 'X'), ('X', 'Y1'), ('W2', 'Y2')],
                    [('W1', 'W2', 'U_W1W2'), ('W1', 'Y2', 'U_W1Y2'), ('W2', 'X', 'U_W2X'), ('W1', 'Y1', 'U_W1Y1')])

try:
    P1 = myID(Y={'Y1', 'Y2'}, X={'X'}, G=G)
except HedgeFound as H:
    print(f"myID Failed: {H._message}")
try:
    P2 = mygID(Y={'Y1', 'Y2'}, X={'X'}, Z=[set()], G=G)
except ThicketFound as T:
    print(f"mygID Failed: {T._message}")
```
